```python
import tkinter
from tkinter import ttk
import pymongo

client = pymongo.MongoClient("mongodb+srv://user1:kang10330154@cluster0.np7oq.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
db = client.gov2
col=db.classes2

window = tkinter.Tk()
window.title("職訓課程查詢")
window.geometry('800x600')
window.iconbitmap('Taiwan.ico')
window.resizable(False, False)
window.config(bg="#00FFFF")
##############################
def label(text, r, c, bg, fg):
    lb = tkinter.Label(window, text=text)
    lb.grid(row=r, column=c, sticky='WE')
    lb.config(font=('微軟正黑體', 15), bg=bg, fg=fg)
##############################
def find_lst(text):
    results = list(col.find())
    lst = list({result[text] for result in results})
    return lst
##############################
def chk(r, c):
    chkValue = tkinter.BooleanVar() 
    chkValue.set(True)
    chk = tkinter.Checkbutton(window, text='全選', var=chkValue, font=('微軟正黑體',15))
    chk.grid(row=r, column=c, sticky='WE')
    return chkValue
##############################
label("訓練性質", 0, 0, "red", "yellow")
lst = find_lst("性質")
combo1 = ttk.Combobox(window, 
                      values=lst,
                      state='readonly',
                      font=('微軟正黑體', 15)
                     )

combo1.grid(row=0, column=1, sticky="WE")
combo1.current(0)

def callback1(event):
     print("新選項")
combo1.bind("<<ComboboxSelected2>>", callback1)

chkValue1 = chk(0, 2)
##############################
label("上課城市", 1, 0, "red", "yellow")

lst = find_lst("上課城市")
combo2 = ttk.Combobox(window, 
                      values=lst,
                      state='readonly',
                      font=('微軟正黑體', 15))

combo2.grid(row=1, column=1, sticky="WE")
combo2.current(0)

def callback2(event):
     print("新選項")
combo2.bind("<<ComboboxSelected2>>", callback2)

chkValue2 = chk(1, 2)
##############################
label("職訓類別", 2, 0, "red", "yellow")

lst = find_lst("職訓類別")
combo3 = ttk.Combobox(window, 
                      values=lst,
                      state='readonly',
                      font=('微軟正黑體', 15))

combo3.grid(row=2, column=1, sticky="WE")
combo3.current(0)
def callback3(event):
     print("新選項")
combo3.bind("<<ComboboxSelected3>>", callback3)


chkValue3 = chk(2, 2)
##############################
label("課程關鍵字", 3, 0, "blue", "yellow")

en1 = tkinter.Entry(window, text="課程關鍵字",font=("微軟正黑體",15))
en1.grid(row=3, column=1, sticky="WE")
##############################
label("結果", 4, 0, "black", "white")

scrollbar = tkinter.Scrollbar(window)
scrollbar.grid(row=4, column=2, sticky="NS")
text1 = tkinter.Text(window, yscrollcommand=scrollbar.set, undo=True)
text1.grid(row=4, column=1, sticky="WE")

scrollbar.config(command=text1.yview)
##############################
def check_filter(chkValue1, chkValue2, chkValue3, col):
    check = [chkValue1.get(), chkValue2.get(), chkValue3.get()]
    key = ["性質", '上課城市', '職訓類別']
    value = [combo1.get(), combo2.get(), combo3.get()]
    dic = {key[x]:value[x] for x in range(len(check)) if not check[x]}
    results = list(col.find(dic))
    return results

def insert_text(result, item):
    if not result[item]:
        result[item] = "其他"
    text1.insert("insert", "%s："%item)
    text1.insert("insert", result[item])
    text1.insert(tkinter.INSERT, '\n')

def insert_text_command(result):
    text1.tag_config("tag_1", backgroun="black", foreground="yellow")
    text1.insert("insert", "課程名稱：", "tag_1")
    text1.insert("insert", result['課程名稱'], "tag_1")
    text1.insert(tkinter.INSERT, '\n')
    lst = ["上課地點", "報名開放時間", "報名截止時間", "課程開始時間", "課程結束時間", "課程時間", "職訓類別", "課程連結"]
    for item in lst:
        insert_text(result, item)
    text1.insert("insert", "-"*50)
    text1.insert(tkinter.INSERT, '\n')
        
def print_item_t(results, t):
    for result in results:
            if t in result['課程名稱']:
                insert_text_command(result)

def print_item(results):
    for result in results:
        insert_text_command(result)

def check():
    t = en1.get()          
    results = check_filter(chkValue1, chkValue2, chkValue3, col)
    if t:
        print_item_t(results, t)
    else:
        print_item(results)

def renew():
    text1.edit_undo()
        
btn1 = tkinter.Button(text="確定", command=check, font=('微軟正黑體', 15), bg='gray65')
btn1.grid(row=5, column=1, sticky="WE")

btn2 = tkinter.Button(text="復原", command=renew, font=('微軟正黑體', 15), bg='gray65')
btn2.grid(row=5, column=2, sticky="WE")

window.mainloop()
```

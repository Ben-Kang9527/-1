```python
import requests
import json
import time
import random
import pymongo

client = pymongo.MongoClient("mongodb+srv://user1:kang10330154@cluster0.np7oq.mongodb.net/myFirstDatabase?retryWrites=true&w=majority")
db = client.gov2
col=db.classes2

i=0
while True:
    url = 'https://course.taiwanjobs.gov.tw/api/Course/paging'
    payload = {
        'Limit': 10,
        'Offset': i,
        'TrainingStartDateYear': 2022,
        'TrainingStartDateMonth': 1,
        'TrainingEndDateYear': 2022,
        'TrainingEndDateMonth': 6,
        'TrainingPeriod': 1,
        'Sort': 'RegisterStartDateTime',
        'Order': 'desc'
    }
    res = requests.post(url, payload)
    
    datas = json.loads(res.text)
    datas = datas['rows']
    
    if not datas:
        break
        
    print("-"*50)
    print("正在讀取第%d頁"%(int(i/10)+1))
    print("-"*50)
    
    for data in datas:
        name = data['Name']
        city_name = data['CityName']
        if city_name == "0":
            city_name = "居家上課"
        address = data['Address']
        if address=="":
            address = city_name
        register_start_datetime = data['RegisterStartDateTime']
        register_end_datetime = data['RegisterEndDateTime']
        training_start_datetime = data['TrainingStartDateTime']
        training_end_datetime = data['TrainingEndDateTime']
        training_time = data['TrainingTime']
        training_unit = data['TrainingUnit']
        category = data['Category']
        course_type_name = data['CourseTypeName']
        class_url = data['Url'].strip()
        col.insert_one({
        "課程名稱":name,
        "上課地點":address,
        "上課城市":city_name,
        "報名開放時間":register_start_datetime,
        "報名截止時間":register_end_datetime,
        "課程開始時間":training_start_datetime,
        "課程結束時間":training_end_datetime,
        "課程時間":training_time,
        "訓練單位":training_unit,
        "職訓類別":category,
        "性質":course_type_name,
        "課程連結":class_url
        })      
        time.sleep(random.randint(1, 5))
    i+=10
print("完成")
```
